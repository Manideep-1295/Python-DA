Key Points:
- filter(items=...): Useful for selecting specific rows or columns by label.
- filter(like=...): Useful for selecting columns or rows that contain a given substring.
- filter(regex=...): Useful for selecting columns or rows that match a regular expression pattern.
- filter(axis=...): Specify whether to filter rows (axis=0) or columns (axis=1).

df.filter(items=None, like=None, regex=None, axis=None).head()

```python
df.filter(items=None, like="a", regex=None, axis=None).head()
```
