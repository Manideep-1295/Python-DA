```python
df1 = read_excel('Customer.xlsx',sheet_name='List')
print(df1.head())
```

```python
print(df1.describe())
```

```python
df2 = df1.drop_duplicates()
df2
```

```python
df2['Phone_Number'] = df2['Phone_Number'].str.replace('N/a','0')
df2
```

```python
df2["Phone_Number"] = df2["Phone_Number"].fillna(0)
df2
```

```python
df2[["Street","State","Zip Code"]] = df2["Address"].str.split(",",n=2,expand = True)
df2
```

```python
df2 = df2.drop(columns=["Address","Not_Useful_Column"])
df2
```

```python
df2["Do_Not_Contact"] = df2["Do_Not_Contact"].fillna(0)
df2
```

```python
df2["Paying Customer"] = df2["Paying Customer"].str.replace("N/a","0")
df2
```

Replacing single character to the required string

```python
# s.apply(func, convert_dtype=True, args=())

df2["Paying Customer"] = df2["Paying Customer"].apply(lambda x: 'Yes' if x == 'Y' else x)
df2["Paying Customer"] = df2["Paying Customer"].apply(lambda x: 'No' if x == 'N' else x)
df2
```
